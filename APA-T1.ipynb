```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
```

```python
T = 2.5
fm = 8000
fx = 1000
A = 4
pi = np.pi
L = int(fm * T)
Tm = 1/fm
t=Tm*np.arange(L) 
x = A * np.cos(2 * pi * fx * t)
sf.write('so_exemple1.wav', x, fm)
```

```python
Tx = 1/fx
Ls = int(fm * 5 * Tx)
plt.plot(t[: Ls], x[: Ls])
```

```python
import sounddevice as sd
# sd.play(x, fm)
```

```python
from numpy.fft import fft
N = 5000
X = fft(x[: Ls], N)
k = np.arange(N)
fig, axs = plt.subplots(2, 1)
axs[0].plot(k, abs(X))
axs[1].plot(k, np.unwrap(np.angle(X)))

fig.suptitle("transformada Fourier del senyal")
axs[0].set_title("Modul")
axs[1].set_title("Fase")
```
